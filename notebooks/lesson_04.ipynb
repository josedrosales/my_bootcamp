```python
#Bernoulli
#El valor esperado de un evento Bernoulli con probabilidad de exito p y no exito q (1-p) es p
#La varianza, desviacion estandar al cuadrado es, p*q
```

```python
n_experiments = 50
probability = 0.5
data = range(n_experiments)
pmf = binom.pmf(data,n_experiments,probability)
plt.figure()
plt.stem(data,pmf)
plt.show()
```

```python
#Poisson
#Esta distribucion representa el numero de eventos que ocurre en un intervalo fijo
```

```python
n_experiments = 10
n_occur = 5
data =  range(n_occur*5)
pmf = poisson.pmf(data,n_occur)
plt.figure()
plt.stem(data,pmf)
plt.show()
```

```python
#Distribucion normal
mu = 0
sigma = 1
data = np.linspace(-5,5,1000)
pdf = norm.pdf(data,mu,sigma)
plt.plot(data,pdf)
plt.show()
```
